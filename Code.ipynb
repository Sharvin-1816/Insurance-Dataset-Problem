```python
categorical_columns = ['Gender','Education Level','Occupation','Location',
                       'Customer Feedback','Marital Status','Exercise Frequency','Smoking Status','Property Type','Policy Type']
```

## Handling Date time

```python
data['Policy Start Date'] = pd.to_datetime(data['Policy Start Date'])
data
```

```python
data['policy_year'] = data['Policy Start Date'].dt.year
```

```python
data['policy_month'] = data['Policy Start Date'].dt.month
```

```python
data['policy_day'] = data['Policy Start Date'].dt.day
```

```python
data
```

```python
data = data.drop(columns=['Policy Start Date'])
```

```python
data
```

```python
numerical_columns = []

for i in data.columns:
    if i not in categorical_columns:
        numerical_columns.append(i)
```

```python
numerical_columns.remove('Premium Amount')
numerical_columns
```

```python
categorical_columns
```

```python
# Ordinal encoding ==> Education level, Policy type, Customer Feedback, Property type, Occupation, Exercise Frequency, Location
# Date time column ===> Policy start date
# One hot encoding ===> Gender, Smoking status, Maternal Status
```

```python
data[categorical_columns] = data[categorical_columns].fillna('missing')
```

## Ordinal encoding

```python
po = [['PhD',"Master's","Bachelor's",'High School'],['Basic', 'Comprehensive','Premium'],['missing','Poor', 'Average','Good'],['Condo', 'Apartment', 'House'],['missing','Employed','Self-Employed','Unemployed'],['Daily','Weekly','Monthly','Rarely'],['Urban', 'Suburban','Rural'],['missing','Married','Divorced','Single'],['No', 'Yes'],['Female', 'Male']]
ordinal_categories = ['Education Level', 'Policy Type', 'Customer Feedback', 'Property Type','Occupation','Exercise Frequency','Location','Marital Status','Smoking Status','Gender']

for i in range(len(po)):
    encoder = OrdinalEncoder(categories=[po[i]])
    data[ordinal_categories[i]] = encoder.fit_transform(data[[ordinal_categories[i]]])
```

```python
data
```

## Normalization

```python
data[numerical_columns] = data[numerical_columns].apply(lambda col: col.fillna(col.mean()))
```

```python
# Normalization

scaler = MinMaxScaler()
premium_scaler = MinMaxScaler()
data[numerical_columns] = scaler.fit_transform(data[numerical_columns])
data['Premium Amount'] = premium_scaler.fit_transform(data[['Premium Amount']])
print(scaler)
```

```python
data
```

```python
X = data.drop(columns=['Premium Amount'])  
y = data['Premium Amount']
```

```python
data
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
rmsle = np.sqrt(mean_squared_log_error(y_test, y_pred))
```

```python
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
print("RMSLE:", rmsle)
```
